```python
from datasets import load_dataset

ds = load_dataset("MSherbinii/mvtec-ad-cable")
```

```python
import os
from PIL import Image

os.makedirs("project_dataset", exist_ok=True)

count = 0

for split in ["train", "test"]:
    for img in ds[split]:

        image = img["image"]

        image.save(f"project_dataset/img_{count}.png")

        count += 1

        if count >= 100:
            break

    if count >= 100:
        break

print("Saved 100 images.")
```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd

DATASET_PATH = "project_dataset"
OUTPUT_PATH = "results"

os.makedirs(OUTPUT_PATH, exist_ok=True)

# Catenary equation
def catenary(x, a, x0, y0):
    return a * np.cosh((x - x0) / a) + y0

results_list = []

for img_name in os.listdir(DATASET_PATH):

    img_path = os.path.join(DATASET_PATH, img_name)

    img = cv2.imread(img_path)

    if img is None:
        continue

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Edge detection
    edges = cv2.Canny(gray, 50, 150)

    # Morphological closing
    kernel = np.ones((5,5), np.uint8)

    morph = cv2.morphologyEx(
        edges,
        cv2.MORPH_CLOSE,
        kernel
    )

    # Find contours
    contours, _ = cv2.findContours(
        morph,
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_NONE
    )

    if len(contours) == 0:
        continue

    # Select largest contour (outer cable boundary)
    largest_contour = max(
        contours,
        key=cv2.contourArea
    )

    points = largest_contour.squeeze()

    if len(points.shape) < 2:
        continue

    x_all = points[:, 0]
    y_all = points[:, 1]

    # Keep only top half points
    height = img.shape[0]

    top_indices = y_all < height * 0.5

    x = x_all[top_indices]
    y = y_all[top_indices]

    if len(x) < 50:
        continue

    try:

        popt, _ = curve_fit(
            catenary,
            x,
            y,
            p0=[50, np.mean(x), np.mean(y)],
            maxfev=5000
        )

        x_fit = np.linspace(
            min(x),
            max(x),
            400
        )

        y_fit = catenary(
            x_fit,
            *popt
        )

        # Normalized error
        error = np.mean(
            np.abs(y - catenary(x, *popt))
        ) / img.shape[0]

        # Sag calculation
        sag = max(y_fit) - min(y_fit)

    except:
        continue

    base = img_name.split(".")[0]

    # Save edges
    cv2.imwrite(
        os.path.join(
            OUTPUT_PATH,
            base + "_edges.png"
        ),
        edges
    )

    # Plot result
    plt.figure(figsize=(6,4))

    plt.imshow(img[:,:,::-1])

    plt.scatter(
        x,
        y,
        s=1,
        color='blue'
    )

    plt.plot(
        x_fit,
        y_fit,
        'r',
        linewidth=2
    )

    plt.axis("off")

    plt.savefig(
        os.path.join(
            OUTPUT_PATH,
            base + "_final.png"
        ),
        bbox_inches='tight'
    )

    plt.close()

    results_list.append({
        "Image": img_name,
        "Mean Error": error,
        "Sag (pixels)": sag
    })

df = pd.DataFrame(results_list)

df.to_csv(
    os.path.join(
        OUTPUT_PATH,
        "results_table.csv"
    ),
    index=False
)

print("Processing Complete")
print(df)
```
